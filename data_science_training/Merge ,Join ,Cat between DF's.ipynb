# Merge operation in pandas

```python
import pandas as pd
player=['mr.x','johan','light','pain']
team=['robary','cold blood','genius','akatsuki']
anime=['unverified','monster','death note','naruto']
df = pd.DataFrame({'charecter':player,'team':team,'anime':anime})
df
```

```python
# creating second DF

rating=[5,9,8,6]
player=['dr.tenma','johan','light','rein']
anime=['unverified','monster','death note','naruto']
df2=pd.DataFrame({'charecter':player,'anime':anime,'rating':rating})
df2
```

```python
# left merge for charecter
# no change in 1st DF
df.merge(df2,on='charecter',how='left')
```

```python
# right merge for charecter
# no change in 2nd DF
df.merge(df2,on='charecter',how='right')
```

```python
# outer merge

df.merge(df2,on='charecter',how='outer')
```

# join operation in Pandas

```python
# the diffrence between join and merge is 
# Merge--> DF merge by Attribute_Names
# Join --> DF join by Index_Name
```

```python
player=['mr.x','johan','light','pain']
team=['robary','cold blood','genius','akatsuki']
anime=['unverified','monster','death note','naruto']
df3 = pd.DataFrame({'charecter':player,'team':team,'anime':anime},index=['a','b','c','d'])
df3
```

```python
player=['mr.x','johan','light','pain']
team=['robary','cold blood','genius','akatsuki']
anime=['unverified','monster','death note','naruto']
df4 = pd.DataFrame({'charecters':player,'teams':team,'animes':anime},['a','b','e','f'])
df4
```

```python
# inner join
df3.join(df4,how='inner')
```

```python
# left join
df3.join(df4,how='left')

```

```python
# right join
df3.join(df4,how='right')
```

```python
# outer join
df3.join(df4,how='outer')
```

# Cat DF's in Pandas

```python
pd.concat([df3,df4])
```
